```python
import matplotlib.pyplot as plt
```

```python
activities = ['eat','sleep','work','play']
slices = [3,7,8,6]
colors = ['r','y','g','b']
plt.pie(slices,labels = activities,colors = colors,startangle = 90,shadow = False,explode = (0,0,0.1,0),radius = 1.2,
        autopct = '%1.1f%%')
plt.legend()
plt.show()
```
