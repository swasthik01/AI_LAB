```python
import pandas as pd
import matplotlib.pyplot as plt
cars_df=pd.read_csv("C:\\Users\\GPT BANTWAL\\Downloads\\Cars931611042729436\\Cars93.csv")
colums=['Manufacturer','Model','Type','Price','MPG.city','MPG.highway','Horsepower','Rear.seat.room','Passengers']
cars_df[colums].head()
```

```python
cars_df["Price"].plot(kind='box',figsize=(4,4))
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(8)
ax1.boxplot(cars_df["Horsepower"])
ax2.boxplot(cars_df["MPG.city"])
ax1.set_title("Horsepower")
ax1.set_xlabel("Horsepower")
ax2.set_title("City Mileage")
ax2.set_xlabel("City Mileage(in miles per US gallon)")
fig.tight_layout()
```
